```python
#H0: In the population, the distribution of likes in trending date 17.14.11 is the same for likes in trending date 17.15.11
#H1:In the population, the distribution of likes in trending date 17.14.11 is less than the likes in trending date 17.15.11
```

```python
#Test Statistic : observation difference between the average of 2 groups
```

```python
means_table  = yt_data.groupby('trending_date').mean()
means_table
```

```python
observed_difference = means_table['likes'][13] - means_table['likes'][15]
observed_difference
```

```python
shuffled = yt_data.sample(7974,replace = False)
shuffled
```

```python
means_table  = shuffled.groupby('trending_date').mean()
means_table
```

```python
diff = means_table['likes'][13] - means_table['likes'][15]
diff
```

```python
differences = np.zeros(5000)
```

```python
for i in np.arange(5000):
  yt_data = df[['trending_date', 'likes']]
  shuffled = yt_data.sample(7974,replace = False)
  shuffled_volume = shuffled['likes']
  original_and_shuffled = yt_data.assign(shuffled_volume=shuffled_volume.values )
  all_group_means= original_and_shuffled.groupby('trending_date').mean()
  difference = all_group_means['shuffled_volume'][13]- all_group_means['shuffled_volume'][15]
  differences[i]=difference
```

```python
differences_df = pd.DataFrame(differences)
differences_df
```

```python
differences_df.hist(bins = 20)
plt.title('Prediction Under Null Hypotheses');
plt.xlabel('Differences between Group Averages',fontsize=15)
plt.ylabel('Units',fontsize=15);
plt.scatter(observed_difference,100,color='red',s=30)
```

```python
P_Value = np.count_nonzero(differences <= observed_difference)/differences.size
P_Value
```

```python
#Since P_Value is large .The conclusion of the test is that the data support the null more than they support the alternative.Hence we fail to reject null hypothesis. It says they have same no. of likes for both the dates and it differs only due to chance
```
